## Seminář obecné psychologie - Inteligence

### Výzkumné cíle: 
- Srovnání výsledků skupiny studentů s dostupnými normami, resp. s výsledky získanými v rámci validizačních studií
- Odhad míry nezávislosti konstruktů měřených jednotlivými testy

### Měřené proměnné: 
- Obecná inteligence
- Emoční inteligence
- Sociální inteligence (Social Information Processing, Social Skills, Social Awareness)
- Kognitivní reflexivita (viz jeden z předchozích seminářů)
- Tendence ne/podléhat kognitivním zkreslením (viz jeden z předchozích seminářů)
- Osobnostní rysy Extraverze, Přívětivosti, Svědomitosti, Emoční stability a Otevřenost ke zkušenosti (viz jeden z předchozích seminářů)

### Metody: 
- 16 PF (Faktor B) (Cattell, 1998)
- [Measure of Emotional Intelligence](https://www.psytoolkit.org/survey-library/emotional-intelligence.html) (Schutte et al., 1998)
- [Tromsø Social Intelligence Scale](https://www.psytoolkit.org/survey-library/social-intelligence-tsis.html) (Silvera, Martinussen, & Dahl, 2001)
- CRT-L2 ([Frederick, 2005](http://emilkirkegaard.dk/en/wp-content/uploads/Shane-Frederick-Cognitive-Re%EF%AC%82ection-and-Decision-Making.pdf); [Primi et al., 2015](https://onlinelibrary.wiley.com/doi/abs/10.1002/bdm.1883); [Toplak, West, & Stanovich, 2014](http://keithstanovich.com/Site/Research_on_Reasoning_files/Toplak_West_Stanovich_14.pdf))
- Sada 6 úloh inspirovaných výzkumem heuristik a zkreslení, které zjišťují některé aspekty racionálního myšlení ([Toplak, West, & Stanovich, 2011](http://www.keithstanovich.com/Site/Research_on_Reasoning_files/MC_2011.pdf))
- [TIPI](https://gosling.psy.utexas.edu/scales-weve-developed/ten-item-personality-measure-tipi/) (Gosling, Rentfrow, & Swann, 2003)

### Inferenční statistika

#### Porovnání průměrné obecné inteligence studentů s normou (*M<sub>2500</sub>* = 9.1, *SD* = 3.4)

```r
# Jednovýběrový t-test
t.test(myData$Obecna_inteligence, mu=9.1)
```

#### Porovnání průměrné emoční inteligence studentů s normou
* *M<sub>ženské vězeňkyně</sub>* = 120
* *M<sub>muži</sub>* = 125
* *M<sub>ženy</sub>* = 131
* *M<sub>terapeuti</sub>* = 135

```r
# Jednovýběrový t-test
t.test(myData$Emocni_inteligence, mu=135)
t.test(myData$Emocni_inteligence, mu=120)
```

#### Porovnání průměrné sociální inteligence studentů s normou


```r
# Jednovýběrový t-test
t.test(myData$Social_information_processing, mu=33.09)
t.test(myData$Social_skills, mu=33.20)
t.test(myData$Social_awareness, mu=37.52)
```

#### Vztah mezi obecnou inteligencí, emoční inteligencí, sociální inteligencí, vybranými aspekty racionality a osobnostními charakteristikami

```r
# Korelační matice
library(corrplot)
col <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))

M <- myData %>%
select(-Vek, -Pohlavi) %>%
cor(use="pairwise.complete.obs", method="pearson")

res <- myData %>%
select(-Vek, -Pohlavi) %>%
cor.mtest(conf.level = .95)

corrplot(M, method = "color", col = col(200),
         type = "upper", order = "AOE", number.cex = .7, # AOE, hclust, FPC
         addCoef.col = "black", # Add coefficient of correlation
         tl.col = "black", tl.srt = 90, # Text label color and rotation
         # Combine with significance
         p.mat = res$p, sig.level = 0.05, insig = "blank", 
         # hide correlation coefficient on the principal diagonal
         diag = FALSE)
```

```r
# Vizualizace korelační matice pomocí metody multidimenzionálního škálování
library(ggpubr)
mds.cor <- (1 - M) %>%
cmdscale() %>%
as_tibble()

colnames(mds.cor) <- c("Dim.1", "Dim.2")
ggscatter(mds.cor, x = "Dim.1", y = "Dim.2", 
          size = 1,
          label = colnames(res.cor),
          repel = TRUE)
```
